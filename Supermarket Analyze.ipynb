### Visualizating data

```python
# Branch 
branch = df['Branch'].value_counts()
branch
```

```python
ax = sns.barplot(x=branch.index, y=branch.values, palette='tab20' )
ax.bar_label(ax.containers[0])
plt.xlabel(' ')
plt.ylabel('Number of people')
plt.title('The number of people from each branch')
plt.show()
```

```python
# Customer type 
customer = df['Customer type'].value_counts()
customer 
```

```python
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99']
```

```python
plt.figure(figsize=(8, 6))
plt.rcParams.update({'font.size': 12})
plt.pie(customer.values,
        labels= customer.index,
        autopct='%1.1f%%',colors=colors,
        pctdistance=0.9)
plt.legend(bbox_to_anchor=(1, 1), loc=2, frameon=False)
plt.title('The component of customer type')
plt.show()
```

```python
# Gender 
gender = df['Gender'].value_counts()
gender
```

```python
plt.figure(figsize=(8, 6))
plt.rcParams.update({'font.size': 12})
plt.pie(gender.values,
        labels= gender.index,
        autopct='%1.1f%%',colors=colors,
        pctdistance=0.9)
plt.legend(bbox_to_anchor=(1, 1), loc=2, frameon=False)
plt.show()
```

```python
# Product Line 
product_line = df.groupby('Product line')['Quantity'].sum().reset_index()
```

```python
colors = colors = ['#FFFF14','#F5F5DC','#13EAc9','#D1B26F','#C79FEF','#7BC8F6']
plt.figure(figsize=(10,8))
ax = sns.barplot(x=product_line['Product line'], y=product_line['Quantity'], palette='tab20' )
ax.bar_label(ax.containers[0])
plt.xlabel(' ')
plt.ylabel('Total Quantity')
plt.title('Total Quantity of Each Product Type')
plt.xticks(rotation=-45)
plt.show()
```

```python
# Payment 
payment = df.groupby('Payment')['Invoice ID'].count().reset_index()
payment
```

```python
plt.figure(figsize=(10,8))
ax = sns.barplot(x=payment['Payment'],y=payment['Invoice ID'], palette='pastel')
ax.bar_label(ax.containers[0])
plt.xlabel(' ')
plt.ylabel('Times')
plt.title('The component of payment')
plt.show()
```
